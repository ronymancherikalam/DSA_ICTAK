```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('iris.xls')
```

# Plots

```python
sns.pairplot(data,hue="Classification")
```

```python
sns.heatmap(data.corr(),annot=True)
# heat maps shows there is high correlation between PL-PW , correlation is high for SL-PL and SL-PW. Regarding -ve correlation none of the variables have high inverse relationship on other
```

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.scatter(x='SL',y='SW',data=data,color='violet')
plt.title('Sepal Comparison',fontsize=16)
plt.xlabel('Sepal Length')
plt.ylabel('sepal Width')

plt.subplot(1,2,2)
plt.scatter(x='PL',y='PW',data=data,color='red')
plt.title('Petal Comparison',fontsize=16)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.tight_layout()


#  As per below plot Some species petals are smaller and sepal distribution need deeper analysis
```

```python
sns.scatterplot(x='PL',y='PW',data=data,hue='Classification')
# Comparitively iris-setosa has a smaller petal Size and largest being Iris-virginica and versicolour comes in the median
```

```python
sns.scatterplot(x='SL',y='SW',data=data,hue='Classification')
# As per below plot we cannot deduce an impact on SL-SW properly and hence -ve correlation is close to 0
```

```python
sns.scatterplot(x='SL',y='PL',data=data,hue='Classification')
# setosa has less length PL and SL since flower is smaller compared to other two
```

```python
sns.scatterplot(x='SL',y='PW',data=data,hue='Classification')
```
